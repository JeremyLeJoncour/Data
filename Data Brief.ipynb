## 3.Traitement des données

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("credit_immo.csv")
data.shape
```

```python
data
```

```python
X=data.iloc[:,-9:-1].values
Y=data.iloc[:,-1].values
```

```python
#Dans la colonne Niv_Etude_Bac
imptr = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imptr.fit(X[:,0:1])
X[:,0:1] = imptr.transform(X[:,0:1])
X[:,0:1]
```

```python
#Dans la colonne enfant_a_Charge
imptr.fit(X[:,7:8])
X[:,7:8] = imptr.transform(X[:,7:8])
X[:,7:8]
```

```python
encoder = LabelEncoder()
X[:,2]=encoder.fit_transform(X[:,2])
X[:,5]=encoder.fit_transform(X[:,5])
Y = encoder.fit_transform(Y)
Y
```

```python
X
```

```python
X_app, X_test, Y_app, Y_test = train_test_split(X,Y,test_size=0.2,random_state=3)
```

### 5.3 Analyse de corrélations

```python
data6 = pd.read_csv("iris.csv")
data6
```

```python
data6.corr()
```

```python
x = data6["largeur_sepal"]
y = data6["longueur_sepal"]
plt.scatter(x, y)
#Pas de corrélation
```

```python
x = data6["longueur_petal"]
y = data6["largeur_petal"]
plt.scatter(x, y)
#Forte Corrélation
```

```python
import seaborn as sb
sb.lmplot(x="longueur_petal", y="largeur_petal", data=data6, fit_reg=True, hue='espèce')
```

```python
x = data6["longueur_petal"]
y = data6["largeur_petal"]

#Corrélation et p-value : Corr proche de 1 ou -1 = Forte corrélation. Corr proche de 0 = pas de corrélation.
from scipy import stats
s.stats.pearsonr(x, y)
```
